```python
# ===============================
# Collect all PKL file names from MotionGPT experiment
# ===============================

from pathlib import Path

# --------- CONFIG ---------
exp_name = "DETO_RVQ_Hierarchical_3layer_youtube"
results_root = Path("results/mgpt")   # change if needed
# --------------------------

exp_dir = results_root / exp_name

if not exp_dir.exists():
    raise FileNotFoundError(f"Experiment folder not found: {exp_dir}")

# collect names
pkl_names = set()

for rank_dir in sorted(exp_dir.glob("test_rank_*")):
    
    if rank_dir.is_dir():
        
        for pkl_file in rank_dir.glob("*.pkl"):
            pkl_names.add(pkl_file.name)

# convert to sorted list
pkl_names = sorted(pkl_names)

# print summary
print(f"Experiment: {exp_name}")
print(f"Total PKL files found: {len(pkl_names)}")

# preview first 10
print("\nFirst 10 files:")
for name in pkl_names[:10]:
    print(name)

# variable available for later use
pkl_names
```
